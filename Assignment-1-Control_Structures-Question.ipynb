```python
kgs = float(input("Enter the body weight in Kgs:"))
if(kgs<18.5):
    print("Underweight")
elif(kgs<=24.9):
    print("Heavy Weight")
elif(kgs<=29.9):
    print("Overweight")
elif(kgs<=34.9):
    print("Obese Class 1")
elif(kgs<=39.9):
    print("Obese Class 2")
else:
    print("Obese Class 3")
```
