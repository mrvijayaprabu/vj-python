```python
#Print Odd Numbers in the list
numbers = (20, 10, 16, 19, 25, 1, 276, 188)
for num in numbers:
    if num % 2 != 0:
        print(num, "is odd")
```

```python
#Print Even numbers in the list
numbers = (20, 10, 16, 19, 25, 1, 276, 188)
for num in numbers:
    if num % 2 == 0:
        print(num, "is even")
```
